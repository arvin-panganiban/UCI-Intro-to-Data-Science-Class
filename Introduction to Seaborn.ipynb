```python
sns.displot(data=iris, x="petal_length", hue="species", multiple="stack")
```

### It is a histogram that shows the distribution of the petal lengths. The x-axis is labeled the petal lengths and the y-axis is the count/quantity of petal lengths. We can conclude that setosas have the smallest petal lengths compared to the other species.

```python
sns.displot(data=iris, x="petal_length", hue="species", multiple="stack", kind="kde")
```

### It is a kernel density plot that shows the density of the petal lengths. The x-axis is labeled the petal lengths and the y-axis the density of the petal lengths. From the graph, we can tell that setosas are more likely to be the same petal length with other setosas.

```python
sns.displot(data=iris, x="petal_length", hue="species", col="species")
```

 ### These are histograms that separate the three species of irises with their own graphs. The x-axis is labeled the petal lengths and the y-axis is the count/quantity of petal lengths. We can determine that virginica have the largest petal lengths compared to the other graphs of the other species from the dataset.

```python
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=iris, x="petal_length", y="sepal_length", hue="species", ax=axs[0])
sns.histplot(data=iris, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

### The left graph represent a scatterplot that compares petal length and sepal length of each irises in the dataset. The x-axis is the petal length while the y-axis is the sepal length. Based on the scatterplot, we can determine that there is a positive correlation between petal length and sepal length for the virginica and versicolor species. However, we do not see a correlation from the setosa species.
### The right graph represents a histogram that counts how many data points we have for each species. The x-axis is the different species of irises while the y-axis is the quantity of each species. We can tell that we have equal amount of data comparing the three species of irises.

```python
sns.jointplot(data=iris, x="petal_length", y="sepal_length", hue="species")
```

### This is a joint disribution of petal and sepal length with marginal axes that shows the univariate distribution of each one separately. The x-axis is the petal length while the y-axis is the sepal length. One thing we can determine from this joint distribution is that most of setosas have 1.5 in petal length and 5 in sepal length.

```python
sns.pairplot(data=iris, hue="species")
```

### This is pairplot that combines joint and marginal views that visualizes every pairwise combination of variables simultaneously. The x-axis and y-axis are the same which are all the quantifiable variables in the dataset. One thing we can conclude from this pairplot is that even though, the petal width of the setosas are the smallest of the two species, they are more likely to have a bigger sepal width compared to the other two irises species.

```python
sns.jointplot(data=iris, x="petal_length", y="sepal_length", hue="species", kind="hist")
```

### This is a jointplot with a kind parameter using histograms. The x-axis is the petal length while the y-axis is the sepal length. One thing we can determine is that most versicolor and virginica irises have the same sepal length.
